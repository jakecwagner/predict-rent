```python
import pandas as pd
import numpy as np
import os

#change working directory
os.chdir("C:/Users/wagne/Dropbox/Rents")

#set pandas defaults for number of rows and columns to display
pd.set_option('display.max_columns', None)
#pd.set_option('display.max_rows', None)

#read in stata data
raw = pd.read_stata("C:/Users/wagne/OneDrive - Washington State University (email.wsu.edu)/WSU_Econ/Research/ahs17.dta")
```

```python
df = raw
#generate a column of constants
df['const'] = 1
#force all data to be float
df = df.astype(float)
#make the control id an integer
df['control'] = df['control'].astype(int)
#browse the data
#df.head()
```

```python
#describe the data
df.describe()
```

```python
#move rent to second column
df = df[ ['control'] + ['weight'] + ['rent'] + [ col for col in df.columns if col != 'control' and col != 'rent' and col != 'weight' ] ]
#drop all the weighting columns
df.drop(df.loc[:,'repweight1':'sp2repwgt160'], inplace = True, axis = 1)
df.drop(df.loc[:,'sp1weight':'sp2weight'], inplace = True, axis = 1)
#drop all flag variables (jvar)
df.drop(df.loc[:,'jacprimary':'jyrbuilt'], inplace = True, axis = 1)
#drop tothcamt (total housing costs) because collinearity
df = df.drop(['tothcamt'] ,axis=1)
#replace rent equal missing if rent=-6
df =df[df['rent']>0]
df.head()
```

```python
#set y equal to outcome variable rent
y = df['rent']
#set x equal to regressors/features (all rows, columns from 'totrooms' to end of dataframe)
x = df.loc[:, 'totrooms':]

#drop utility costs from the set of covariates (could omit as regressors and then define our treatment and control groups strictly)
#x.drop(x.loc[:,'elecamt':'utilamt'], inplace=True, axis=1)
#alternatively just drop wateramt (include others as regressors, and define our treatment/control groups more loosley)
x = x.drop(['wateramt'], axis=1)

#we want to train on the control group, and then predict out of sample for the treatment group
#to do inference, we can pick a pseudo treatment group, and compare our difference between prediction and outcome in the pseudo treated to the difference between prediction and outcome in the true treated group (similar to synthetic control) 
#what should the control group be (could be several)? options: tenant-pays for all utilities, tenant-pays for utility (or set of utilities) in question 
# treatment group indicates they pay for water
control = (df['wateramt']>3) #& (df['elecamt']!=2) & (df['gasamt']!=2) #&(df['oilamt']!=2)#&(df['otheramt']!=2)#&(df['trashamt']!=2)
y_control = y[control]
x_control = x[control]
# treatment group indicates that water is included in rent, site rent, condo fee, other fee, or other utility bill (problematic if in other utility bill)
treated = (df['wateramt']==2) #& (df['elecamt']!=2) & (df['gasamt']!=2)
y_treated = y[treated]
x_treated = x[treated]
```

```python
#run a simple OLS model 
import statsmodels.api as sm

regTest = sm.OLS(endog=y_control, exog=x_control, missing='drop')
resultsTest = regTest.fit()
print(resultsTest.summary())
```

```python
from sklearn.model_selection import ShuffleSplit, train_test_split

# Shuffle and split the data into training and testing subsets
x_train, x_test, yTrain, yTest = train_test_split(x_control, y_control, test_size=0.2)

#use preprocessing to create 'polynomial features' (higher order variables and interactions)
from sklearn import preprocessing
from sklearn.feature_selection import SelectKBest, f_regression
import warnings
warnings.filterwarnings(action='ignore', category=RuntimeWarning) #ignore RuntimeWarning
warnings.filterwarnings(action='ignore', category=FutureWarning)

#find the most important features
selector = SelectKBest(f_regression, k=50)
selector.fit(x_test, yTest)
#use a mask to later call the most important features
mask = selector.get_support()
#save an array of the most important var names
chosenCols = x_control.columns[mask]
print('The best feautres are:', chosenCols)
#define xBestFeatures to be the x matrix include the best features
xBestFeatures =x_train[chosenCols]

#Now we want to take those features, and allow for flexible interactions (probably not necessary in most machine learning applications)
#take the most important features and construct 2nd order polynomials (can't do this for all data because of memory)
poly = preprocessing.PolynomialFeatures(2, interaction_only=True, include_bias=False)
xBestInt = poly.fit_transform(xBestFeatures) 

#merge the data of the best X interactions with the original data
xTrain = pd.merge(x_train, pd.DataFrame(xBestInt).set_index(x_train.index), left_index=True, right_index=True)

##Need to mimic this for the test group
xBestTest = x_test[chosenCols]
xBestIntTest = poly.fit_transform(xBestTest) 
xTest = pd.merge(x_test, pd.DataFrame(xBestIntTest).set_index(x_test.index), left_index=True, right_index=True)


##Need to mimic this for the treated group...
xBestTreated = x_treated[chosenCols]
xBestIntTreated = poly.fit_transform(xBestTreated) 
xTreated = pd.merge(x_treated, pd.DataFrame(xBestIntTreated).set_index(x_treated.index), left_index=True, right_index=True)


## now we have three matrices with the full data set: one for the treated, and two for the controls (train and test)
```

```python
#now pick a subset of the best features to use in analysis
def bestCols(k):
    selector = SelectKBest(f_regression, k=k)
    selector.fit(xTest, yTest)
    return xTest.columns[selector.get_support()]

smallest = bestCols(150)
smaller = bestCols(300)
small = bestCols(500)

xTestSmallest =xTest[smallest]
xTrainSmallest =xTrain[smallest]
xTreatedSmallest =xTreated[smallest]

xTestSmaller =xTest[smaller]
xTrainSmaller =xTrain[smaller]
xTreatedSmaller =xTreated[smaller]

xTestSmall =xTest[small]
xTrainSmall =xTrain[small]
xTreatedSmall =xTreated[small]

#it is often usefull to standardize/scale our data

xscaler = preprocessing.StandardScaler().fit(xTrain)
xTrainScaled = xscaler.transform(xTrain)
xTestScaled = xscaler.transform(xTest)
xTreatedScaled = xscaler.transform(xTreated)

xscalerSmall = preprocessing.StandardScaler().fit(xTrainSmall)
xTrainSmallScaled = xscalerSmall.transform(xTrainSmall)
xTestSmallScaled = xscalerSmall.transform(xTestSmall)
xTreatedSmallScaled = xscalerSmall.transform(xTreatedSmall)

xscalerSmaller = preprocessing.StandardScaler().fit(xTrainSmaller)
xTrainSmallerScaled = xscalerSmaller.transform(xTrainSmaller)
xTestSmallerScaled = xscalerSmaller.transform(xTestSmaller)
xTreatedSmallerScaled = xscalerSmaller.transform(xTreatedSmaller)

xscalerSmallest = preprocessing.StandardScaler().fit(xTrainSmallest)
xTrainSmallestScaled = xscalerSmallest.transform(xTrainSmallest)
xTestSmallestScaled = xscalerSmallest.transform(xTestSmallest)
xTreatedSmallestScaled = xscalerSmallest.transform(xTreatedSmallest)

# # consider standardizing outcome variable
# # reshape 1d arrays to 2d arrays
# y_train = np.array(y_train).reshape(len(y_train), 1)
# y_test = np.array(y_test).reshape(len(y_test), 1)
# yscaler = preprocessing.StandardScaler().fit(y_train)
# y_trainScaled = yscaler.transform(y_train)[:,0] #scale and reshape
# y_testScaled = yscaler.transform(y_test)[:,0]
# #reshpe 2d arrays to 1d arrays
# y_train = y_train.ravel()
# y_test = y_test.ravel()
```

```python
#set up framework for model selection (tuning parameters)
from time import time
from scipy.stats import randint, uniform, norm

from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import GridSearchCV

# Utility function to report best scores
def report(results, n_top=10):
    for i in range(1, n_top + 1):
        candidates = np.flatnonzero(results['rank_test_score'] == i)
        for candidate in candidates:
            print("Model with rank: {0}".format(i))
            print("Mean validation score: {0:.3f} (std: {1:.3f})".format(
                  results['mean_test_score'][candidate],
                  results['std_test_score'][candidate]))
            print("Parameters: {0}".format(results['params'][candidate]))
            print("")
    #need to make a ranking
    n=0
    for i in results['rank_test_score']:
        if i==1:
            top=n
        n=n+1
    return top

            

```

```python
#decision tree regression
from sklearn import tree

#random tree model
treeTest = tree.DecisionTreeRegressor()

# specify parameters and distributions to sample from model selection
param_dist = {"max_depth": randint(1,10000),
              "min_samples_split": randint(2, 50),
              "min_samples_leaf": randint(1,500)}

# run randomized search
n_iter_search=500
random_search = RandomizedSearchCV(treeTest, param_distributions=param_dist,
                                   n_iter=n_iter_search, cv=3, iid=False)

start = time()
random_search.fit(xTrainScaled, yTrain)
print("RandomizedSearchCV took %.2f seconds for %d candidate"
      " parameter settings." % ((time() - start), n_iter_search))
opt = report(random_search.cv_results_)

max_depthOpt = random_search.cv_results_['params'][opt]['max_depth']
min_samples_leafOpt = random_search.cv_results_['params'][opt]['min_samples_leaf']
min_samples_splitOpt = random_search.cv_results_['params'][opt]['min_samples_split']
```

```python
#run the decision tree regressor using optimal hyperparameters

treeOpt = tree.DecisionTreeRegressor(max_depth= max_depthOpt, min_samples_leaf= min_samples_leafOpt, min_samples_split= min_samples_splitOpt) #use optimal hyperparameters from above
treeOpt.fit(xTrainScaled, yTrain)

yPredTreeOpt = treeOpt.predict(xTestScaled)
yPredTreeOptSample = treeOpt.predict(xTrainScaled)
yPredTreeOptTreated = treeOpt.predict(xTreatedScaled)

# The mean squared error
print("Mean squared error (in-sample): %.2f"
      % mean_squared_error(yTrain, yPredTreeOptSample))
# Explained variance score: 1 is perfect prediction
print('R-square (in-sample): %.2f' % r2_score(yTrain, yPredTreeOptSample))
print("Mean squared error: %.2f"
      % mean_squared_error(yTest, yPredTreeOpt))
print('R-square: %.2f' % r2_score(yTest, yPredTreeOpt))

print()
print("Predicted rent (control group (test)): ", np.mean(yPredTreeOpt))
print("Predicted rent (treatment group): ", np.mean(yPredTreeOptTreated))
print("Actual rent (control group (test))", np.mean(yTest))
print("Actual rent (treatment group)", np.mean(y_treated))
print("Diff:", np.mean(y_treated - yPredTreeOptTreated))
print("Average cost", np.mean(df[control]['wateramt'])) #could also predict what wateramt should be for comparison
```

```python
def dist(n):
    array = []
    for i in range(n):
        array.append(round(min(np.random.lognormal(3,1),250)))
    return array
layers=[]
for i in range(100000):
    layers.append(dist(1))
    layers.append(dist(2))
    layers.append(dist(3))
    layers.append(dist(4))
    layers.append(dist(5))
    layers.append(dist(6))
    layers.append(dist(10))
```

```python
#Neural network regression
from sklearn.neural_network import MLPRegressor
from sklearn.exceptions import FitFailedWarning, ConvergenceWarning

warnings.filterwarnings(action='ignore', category=FitFailedWarning) #don't print the FitFailedWarning
warnings.filterwarnings(action='ignore', category=ConvergenceWarning)

neuralTest= MLPRegressor(max_iter=1000,
              early_stopping= True)

# specify parameters and distributions to sample from model selection
param_dist = {"learning_rate_init": uniform(0.00001,0.01),
              "learning_rate": ['constant', 'invscaling', 'adaptive'],
              "power_t": uniform(0,1),
              "solver": ['adam','sgd'],
              "batch_size": randint(31,512),
              "momentum": uniform(0,1),
              "hidden_layer_sizes": layers,
              "nesterovs_momentum":[True,False],
              "alpha": uniform(0.000001, 0.1)}

# run randomized search
n_iter_search=300
random_search = RandomizedSearchCV(neuralTest, param_distributions=param_dist,
                                   n_iter=n_iter_search, error_score=np.nan, cv=3, iid=False) 

start = time()
random_search.fit(xTrainScaled, yTrain)
print("RandomSearchCV took %.2f seconds for %d candidates"
      " parameter settings." % ((time() - start), n_iter_search))
opt = report(random_search.cv_results_)

learning_rate_initOpt = random_search.cv_results_['params'][opt]['learning_rate_init']
learning_rateOpt = random_search.cv_results_['params'][opt]['learning_rate']
power_tOpt = random_search.cv_results_['params'][opt]['power_t']
solverOpt = random_search.cv_results_['params'][opt]['solver']
batch_sizeOpt = random_search.cv_results_['params'][opt]['batch_size'] 
momentumOpt = random_search.cv_results_['params'][opt]['momentum']
hidden_layer_sizesOpt = random_search.cv_results_['params'][opt]['hidden_layer_sizes']
nesterovs_momentumOpt = random_search.cv_results_['params'][opt]['nesterovs_momentum']
alphaOpt = random_search.cv_results_['params'][opt]['alpha']
```

```python
#run the neural net using optimal hyperparmaters

neuralOpt = MLPRegressor(max_iter=1000, early_stopping= True, batch_size= batch_sizeOpt, hidden_layer_sizes= hidden_layer_sizesOpt, learning_rate= learning_rateOpt, learning_rate_init= learning_rate_initOpt, momentum= momentumOpt, nesterovs_momentum= nesterovs_momentumOpt, power_t= power_tOpt, solver= solverOpt, alpha= alphaOpt) 
neuralOpt.fit(xTrainScaled,yTrain)
yPredNeuralOpt =neuralOpt.predict(xTestScaled)
yPredNeuralOptSample = neuralOpt.predict(xTrainScaled)
yPredNeuralOptTreated= neuralOpt.predict(xTreatedScaled)

# The mean squared error
print("Mean squared error(in-sample): %.2f"
      % mean_squared_error(yTrain, yPredNeuralOptSample))
# Explained variance score: 1 is perfect prediction
print('R-square (in-sample): %.2f' % r2_score(yTrain, yPredNeuralOptSample))
print("Mean squared error: %.2f"
      % mean_squared_error(yTest, yPredNeuralOpt))
print('R-square: %.2f' % r2_score(yTest, yPredNeuralOpt))

print()
print("Predicted rent (control group (test)): ", np.mean(yPredNeuralOpt))
print("Predicted rent (treatment group): ", np.mean(yPredNeuralOptTreated))
print("Actual rent (control group (test))", np.mean(yTest))
print("Actual rent (treatment group)", np.mean(y_treated))
print("Diff:", np.mean(y_treated - yPredNeuralOptTreated))
print("Average cost", np.mean(df[control]['wateramt']))
```
